#Dropping unnecessary feaures

```python
data1=data.drop(["ID#"],axis=1)
```

```python
data1
```

#Renaming the features

```python
data2=data1.rename(columns={'Qual_miles':'Qual','cc1_miles':'cc1','cc2_miles':'cc2','cc3_miles':'cc1','Flight_trans_12':'FT','Days_since_enroll':'Days','Award?':'Award','Bonus_miles':'BM','Bonus_trans':'BT','Flight_miles_12mo':'FM'})
data2
```

```python
array=data2.values
array
```

```python
array
```

```python
stscaler = StandardScaler()
X = stscaler.fit_transform(array)
```

```python
X
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=3)
dbscan.fit_predict(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
clustered = pd.concat([data2,cl],axis=1)
```

```python
clustered
```

```python
clustered[clustered['cluster']==-1]
```

so there are 1228 rows contains outliers

```python
clustered
```

```python
clustered.mean()
```
